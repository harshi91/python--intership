```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV file
data = pd.read_csv("sales.csv")

# Display first 5 rows
print("Preview of dataset:")
data.head()
```

```python
# Check info about dataset
data.info()

# Get summary statistics
data.describe()

# Check for missing values
data.isnull().sum()
```

```python
# Total sales by region
region_sales = data.groupby("Region")["Sales"].sum()
print(region_sales)

# Total sales by product
product_sales = data.groupby("Product")["Sales"].sum()
print(product_sales)
```

```python
region_sales.plot(kind='bar', color='skyblue', title='Total Sales by Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
product_sales.plot(kind='bar', color='lightgreen', title='Total Sales by Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Convert Date to datetime
data["Date"] = pd.to_datetime(data["Date"])

# Group by Date
daily_sales = data.groupby("Date")["Sales"].sum()

daily_sales.plot(kind='line', figsize=(8,4), title='Daily Sales Trend')
plt.ylabel('Sales')
plt.show()
```
